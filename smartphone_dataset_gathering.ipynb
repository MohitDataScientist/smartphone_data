```python
# open smartrprix html file in reading mode
with open ('smartprix.html','r',encoding = 'utf-8') as f:
    html = f.read()
```

```python
# imporl all required library 
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
# structure of container of every phone
containers = soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
name = []
price = []
rating = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []
specs_score = []

for i in containers:
       
    # name
    try:
        name.append(i.find('h2').text)
    except:
        name.append(np.nan)
    
    # price
    try:
        price.append(i.find('span',{'class':'price'}).text)
    except:
        price.append(np.nan)
    
   # rating
    try:
        rating.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        rating.append(np.nan)
        
    x = i.find('ul',{'class':'sm-feat specs'}).find_all('li')
    # sim
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
        
    # processor
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
        
    # ram
    try:    
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    
    # battery
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
        
    # display
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
        
    # camera
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
        
    # card
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    
    # os(operating_system)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)
    
    
     # specification_score
    span = (i.find('span', class_='sm-rating'))
    if span:
    # Extract the rating value from the style attribute
        style_attribute = span['style']
        rate = float(style_attribute.split(': ')[1][:-1])
    
    try:
        specs_score.append(rate)
    except:
        specs_score.append(np.nan)
        
        
            
    
    
    
    
    
    
    
    
    
```

```python
specs_score
```

```python
smartphone= pd.DataFrame({
    'model':name,
    'price':price,
    'rating':rating,
    'sim':sim,
    'processor':processor,
    'ram':ram,
    'battery':battery,
    'display':display,
    'camera':camera,
    'card':card,
    'os':os,
    'specs_score':specs_score
})
```

```python
smartphone.to_csv('smartphone.csv',index = False)
```
